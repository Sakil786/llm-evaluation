```python
from beyondllm import source, embeddings, retrieve, llms, generator

data = source.fit(
    path="https://www.youtube.com/watch?v=oJJyTztI_6g",
    dtype="youtube",
    chunk_size=1024,
    chunk_overlap=0)
```

```python
HF_TOKEN="Hugging face Key"
```

```python
# %pip install llama-index-embeddings-huggingface
# %pip install llama-index-embeddings-instructor
```

```python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

# loads BAAI/bge-small-en
# embed_model = HuggingFaceEmbedding()

# loads BAAI/bge-small-en-v1.5
embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en-v1.5")
```

```python
retriever = retrieve.auto_retriever(
data=data,
embed_model=embed_model,
type="cross-rerank",
mode="OR",
top_k=2)
```

```python
llm = llms.HuggingFaceHubModel(
model="mistralai/Mistral-7B-Instruct-v0.3",
token=HF_TOKEN
)
```

```python
system_prompt = f"""
<s>[INST]
You are an AI Assistant.
Please provide direct answers to questions.
[/INST]
</s>
"""

pipeline = generator.Generate(
question=" summarize the content from the video?",
retriever=retriever,
system_prompt=system_prompt,
llm=llm)
```

```python
print(pipeline.call())
```
